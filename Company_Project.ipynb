```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median())) #Ye code har Category ke andar Review Rating ke missing values ko usi category ke median rating se fill karta hai.
```

```python
df.isnull().sum()
```

# Hum Columns name ko lower case me krege or space ko (_) underscore se fill krege. Because its easy to use in coding

```python
df.columns=df.columns.str.lower() # Convert columns name in lower case
df.columns=df.columns.str.replace(' ','_') # fill columns space with underscore

df=df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}) # Rename krne k leye name
```

```python
df.columns # To see columns
```

```python
# Create a Column age_group

labels=['Young Adult','Adult','Middle-aged','Senior'] # jo hume age group k hisab se name dene h
df['age_group']=pd.qcut(df['age'], q=4,labels=labels)  # age column ko 4 equal-sized age groups (quartiles) me divide karta hai aur unko readable labels deta hai 
```

```python
df[['age','age_group']].head(10) # muje age and age group column k top 10 row dekhne k leye
```

```python
# Create a column purchase_frequency_days
df.groupby('frequency_of_purchases')['frequency_of_purchases'].count() # ye dekhne k leye column me kithne count h 

frequency_mapping={      # humne ek Dictionary banaye jo  text ko number me batyegi jaise week me 7 dyas    
'Bi-Weekly' : 14,     
'Every 3 Months': 90,   
'Fortnightly' : 14,      
'Monthly':30,           
'Quarterly' : 90,        
'Weekly' :7, 
 'Annually': 365
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping) # map ka use kiya text ko numnber me replace krne k leye
```
